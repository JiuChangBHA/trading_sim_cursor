```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python

# For equity curve
def plot_equity_curve(file_paths):
    plt.figure(figsize=(12, 6))
    for file_path in file_paths:
        df = pd.read_csv(file_path)
        df['Date'] = pd.to_datetime(df['Date'])
        plt.plot(df['Date'], df['Equity'], label=os.path.basename(file_path))
    plt.legend()
    plt.title('Equity Curve Over Time')
    plt.xlabel('Date')
    plt.ylabel('Equity ($)')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# For metric history
def plot_metric_history(file_path, metric_name):
    df = pd.read_csv(file_path)
    df['Date'] = pd.to_datetime(df['Date'])
    
    plt.figure(figsize=(12, 6))
    sns.lineplot(data=df, x='Date', y=metric_name, hue='Symbol')
    plt.title(f'{metric_name} Over Time by Symbol')
    plt.xlabel('Date')
    plt.ylabel(metric_name)
    plt.grid(True)
    plt.tight_layout()
    plt.show()


# For optimization results
def plot_optimization_results(file_path):
    df = pd.read_csv(file_path)
    df['Date'] = pd.to_datetime(df['Date'])
    
    plt.figure(figsize=(12, 6))
    sns.lineplot(data=df, x='Date', y='Metric', hue='Symbol')
    plt.title(f'Optimization Results Over Time by Symbol')
    plt.xlabel('Date')
    plt.ylabel('Metric')
    plt.grid(True)
    plt.tight_layout()
    plt.show()


# For strategy comparison
def plot_strategy_comparison(file_paths):
    plt.figure(figsize=(12, 6))
    for file_path in file_paths:
        df = pd.read_csv(file_path)
        df['Date'] = pd.to_datetime(df['Date'])
        plt.plot(df['Date'], df['Equity'], label=os.path.basename(file_path))
    
    plt.title('Strategy Comparison')
    plt.xlabel('Date')
    plt.ylabel('Equity ($)')
    plt.grid(True)

def main():
    # Example usage
    equity_curve_file = '../simulation_results/AAPL_Bollinger_Bands_2025-03-10.csv'
    metric_history_file = 'src/main/resources/optimization_reports/AAPL_Moving Average Crossover_optimization_2025-03-14.csv'
    optimization_results_file = 'src/main/resources/optimization_reports/AAPL_Moving Average Crossover_optimization_2025-03-14.csv'
    strategy_comparison_files = [
        'src/main/resources/simulation_results/AAPL_Moving Average Crossover_optimization_2025-03-14.csv',
        'src/main/resources/simulation_results/AAPL_Moving Average Crossover_optimization_2025-03-14.csv'
    ]


    # Plot equity curve
    plot_equity_curve(equity_curve_file)

    # Plot metric history
    plot_metric_history(metric_history_file, 'Sharpe Ratio')

    # Plot optimization results
    plot_optimization_results(optimization_results_file)

    # Plot strategy comparison
    plot_strategy_comparison(strategy_comparison_files)

```

```python
BB_equity_curve_file = '../simulation_results/AAPL_Bollinger_Bands_2025-03-10.csv'
MR_equity_curve_file = '../simulation_results/AAPL_Mean_Reversion_2025-03-10.csv'
RSI_equity_curve_file = '../simulation_results/AAPL_Relative_Strength_Index_(RSI)_2025-03-10.csv'

equity_files = [BB_equity_curve_file, MR_equity_curve_file, RSI_equity_curve_file]
plot_equity_curve(equity_files)
```
